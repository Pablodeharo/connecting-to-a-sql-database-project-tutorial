```python
import psycopg2


# Conectar a la base de datos
conn = psycopg2.connect(
    host='localhost',
    port=5432,  # Puerto por defecto para PostgreSQL
    user='gitpod',  # Nombre de usuario
    password='postgres',  # Contraseña
    database='db_name'  # Nombre de la base de datos
)

# Crear un cursor para ejecutar consultas SQL
cursor = conn.cursor()

# Leer el contenido de cada archivo SQL y ejecutarlo
sql_files = ['/workspaces/connecting-to-a-sql-database-project-tutorial/src/sql/create.sql', 
             '/workspaces/connecting-to-a-sql-database-project-tutorial/src/sql/insert.sql', 
             '/workspaces/connecting-to-a-sql-database-project-tutorial/src/sql/drop.sql']
for file_name in sql_files:
    with open(file_name, 'r') as sql_file:
        sql_script = sql_file.read()
        cursor.execute(sql_script)

# Guardar los cambios
conn.commit()

# Cerrar el cursor y la conexión
cursor.close()
conn.close()

# Mensaje de confirmación
print("Todos los scripts SQL han sido ejecutados con éxito.")
```
